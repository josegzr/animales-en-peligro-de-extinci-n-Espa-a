```python
import pandas as pd 
import numpy as np
```

### 1. cargar el csv de la poblacion de lobos y limpiarlo

```python
df = pd.read_csv('../Data/manadas-de-lobos-por-region.csv')
df
```

```python
df['Parámetro;Exclusivas en la región;Compartidas con otras regiones;Total regional '] = df['Parámetro;Exclusivas en la región;Compartidas con otras regiones;Total regional '].str.split(';')
df = df['Parámetro;Exclusivas en la región;Compartidas con otras regiones;Total regional '].apply(pd.Series)
```

```python
df
```

```python
df.columns = ['Region_lob', 'exclusivas_en_la_region', 'compartidas_con_otras_regiones', 'total']
```

```python
df
```

#### segun un estudio de la fundacion artemisan las manadas de lobos en españa suelen tener entre 7-8 lobos, por lo que voy a multiplicar el total de manadas por 7 para tener el total de lovos por region

```python
df.info()
```

```python
df['exclusivas_en_la_region'] = pd.to_numeric(df['exclusivas_en_la_region'], errors='coerce')
df['compartidas_con_otras_regiones'] = pd.to_numeric(df['compartidas_con_otras_regiones'], errors='coerce')
df['total'] = pd.to_numeric(df['total'], errors='coerce')
```

```python
df['total_lobos'] = df['total'] * 7
```

```python
df
```

```python
df['Especie_lob'] = 'Lobo'
```

```python
df
```

```python
column_order = ['Region_lob','Especie_lob', 'exclusivas_en_la_region', 'compartidas_con_otras_regiones', 'total']
```

```python
df = df[column_order]
```
