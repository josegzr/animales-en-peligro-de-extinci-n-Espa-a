# poblacion de lince iberico, repartido por localidades en España

```python
import pandas as pd
from selenium import webdriver
import time
import warnings
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

```python
url= 'https://lifelynxconnect.eu/censos/'
```

```python
PATH = 'driver/ChromeDriver.exe'

driver = webdriver.Chrome()  

driver.get(url)
driver
```

```python
driver.find_element(By.CSS_SELECTOR, '#cn-accept-cookie').click()
```

```python
tabla = driver.find_element(By.XPATH, '//*[@id="table_1_wrapper"]')

tabla
```

```python
cuerpo = tabla.find_element(By.TAG_NAME, 'tbody')
filas = cuerpo.find_elements(By.TAG_NAME, 'tr')
```

```python
matriz = []


for f in filas:
    
    tmp = []
    
    elems = f.find_elements(By.TAG_NAME, 'td')  # esto es una lista
    
    for e in elems:
        
        tmp.append(e.text)
        
    matriz.append(tmp)
```

```python
cabecera = tabla.find_element(By.TAG_NAME, 'thead')
columnas=[e.text for e in cabecera.find_elements(By.TAG_NAME, 'th')]
columnas
```

```python
df = pd.DataFrame(matriz, columns=columnas)
df
```

##### 2. Quiero saber cuantos machos y hembras no reproductoras hay por lo que hago una funcion para sumar las columnas de cachorros y hembras y restarselo al total. Para llegar a esto tambien tengo que convertir a numerico los valores de las columnas

```python
fila = 5
columna = 'Cachorros'

df.at[fila, columna] = '1'
```

```python
df['Hembras reproductoras'] = pd.to_numeric(df['Hembras reproductoras'], errors='coerce')
df['Cachorros'] = pd.to_numeric(df['Cachorros'], errors='coerce')
df['Total ejemplares'] = pd.to_numeric(df['Total ejemplares'], errors='coerce')
```

```python
df['machos/hembaras_no_repro'] =df['Total ejemplares'] - (df['Hembras reproductoras'] + df['Cachorros'] )
```

```python
df
```

```python
df['Cachorros'] = df['Cachorros'].astype('int64')
df['machos/hembaras_no_repro'] = df['machos/hembaras_no_repro'].astype('int64')
```

```python
column_order = ['Subpoblación', 'Región', 'Hembras reproductoras','Cachorros','machos/hembaras_no_repro','Total ejemplares']
df = df[column_order]
```
